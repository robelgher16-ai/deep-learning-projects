# AMZN Stock Price Prediction with LSTM

This notebook trains a stacked LSTM (Long Short-Term Memory) neural network in PyTorch to predict Amazon's next-day closing price from the previous 7 days of closing prices.

*Fixes applied to the original notebook:*
- Removed a stray leading space before `split_index = ...` that caused an `IndentationError`.
- Removed a broken duplicate plotting cell that referenced undefined variables (`index_with_lagged_dropped`, `unscaled_y`, `full_predictions`).
- Removed 3 exact-duplicate plotting cells at the end (kept the best-labeled one).
- Uncommented the axis labels/title on the raw price plot.
- Added explanatory comments to every meaningful line.

No modeling logic was changed — same LSTM architecture, same training loop, same data pipeline.

```python
# Core libraries
import pandas as pd            # data loading & tabular manipulation
import numpy as np             # numerical arrays
import matplotlib.pyplot as plt  # plotting

import torch                   # PyTorch core (tensors, autograd)
import torch.nn as nn          # neural network building blocks (layers, losses)
```

```python
# Load the raw AMZN price history from CSV
data = pd.read_csv('AMZN.csv')

data  # display the dataframe (Date, Open, High, Low, Close, Adj Close, Volume)
```

```python
# We only need the Date and the Close price for this univariate LSTM
data = data[['Date', 'Close']]
data
```

```python
# Use the GPU if one is available, otherwise fall back to CPU
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
device
```

```python
# Convert the Date column from string to an actual datetime type
data['Date'] = pd.to_datetime(data['Date'])

# Quick sanity-check plot of the raw closing price over time
plt.plot(data['Date'], data['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Amazon Closing Price')
plt.show()
```

```python
from copy import deepcopy as dc  # deepcopy so we never mutate the caller's dataframe
import pandas as pd

def prepare_dataframe_for_lstm(df, n_steps):
    """
    Turn a single 'Close' column into a supervised-learning table:
    for every row, add n_steps extra columns holding the previous
    n_steps closing prices (Close(t-1), Close(t-2), ... Close(t-n_steps)).
    This is the standard "windowing" trick used to feed a time series
    into a model that expects fixed-size inputs.
    """
    df = dc(df)                       # work on a copy, not the original
    df['Date'] = pd.to_datetime(df['Date'])   # ensure Date is datetime

    df.set_index('Date', inplace=True)  # Date becomes the index (not a feature)

    for i in range(1, n_steps + 1):
        # shift(i) pulls the Close value from i rows earlier into this row
        df[f'Close(t-{i})'] = df['Close'].shift(i)

    df.dropna(inplace=True)  # first n_steps rows have no history -> drop them

    return df

lookback = 7  # how many previous days the model gets to see
shifted_df = prepare_dataframe_for_lstm(data, lookback)
shifted_df
```

```python
# Convert the dataframe (Close, Close(t-1) ... Close(t-7)) to a plain numpy array
shifted_df_as_np = shifted_df.to_numpy()
shifted_df_as_np
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scale every column into the range [-1, 1]. LSTMs train much better on
# scaled inputs, and MinMaxScaler is a common choice for price series.
scaler = MinMaxScaler(feature_range=(-1, 1))
shifted_df_as_np = scaler.fit_transform(shifted_df_as_np)

shifted_df_as_np
```

```python
# Column 0 is the target (today's Close); columns 1: are the 7 lagged features
X = shifted_df_as_np[:, 1:]   # inputs: Close(t-1) ... Close(t-7)
y = shifted_df_as_np[:, 0]    # target: Close(t)

X.shape, y.shape
```

```python
# Chronological train/test split: last 5% of the timeline is held out for testing.
# NOTE: fixed a stray leading space here that caused an IndentationError.
split_index = int(len(X) * 0.95)

X_train = X[:split_index]
X_test = X[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Convert numpy arrays to torch tensors.
# unsqueeze(-1) adds a "feature dimension" of size 1, since each of the 7
# timesteps only has ONE feature (the closing price). LSTM expects input
# shape (batch, sequence_length, num_features).
X_train = torch.tensor(X_train).float().unsqueeze(-1)
y_train = torch.tensor(y_train).float()
X_test = torch.tensor(X_test).float().unsqueeze(-1)
y_test = torch.tensor(y_test).float()

# Verify the shapes! X_train should now be 3D: (num_samples, lookback, 1)
print(X_train.shape, X_test.shape)
```

```python
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    """Thin wrapper so PyTorch's DataLoader can iterate over (X, y) pairs."""
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)          # total number of samples

    def __getitem__(self, i):
        return self.X[i], self.y[i]  # one (window, target) pair


train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)
```

```python
from torch.utils.data import DataLoader

batch_size = 16

# shuffle=True for training so the model doesn't learn a fixed batch order
# shuffle=False for testing so predictions stay in chronological order
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Peek at one batch to confirm shapes before building/training the model
for _, batch in enumerate(train_loader):
    x_batch, y_batch = batch[0].to(device), batch[1].to(device)
    print(x_batch.shape, y_batch.shape)
    break
```

```python
import torch.nn as nn

class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_stacked_layers):
        super().__init__()
        self.hidden_size = hidden_size
        self.num_stacked_layers = num_stacked_layers
        # batch_first=True -> input/output tensors shaped (batch, seq, feature)
        self.lstm = nn.LSTM(input_size, hidden_size, num_stacked_layers,
                            batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)  # maps final hidden state -> 1 predicted price

    def forward(self, x):
        batch_size = x.size(0)

        # Initial hidden state (h0) and cell state (c0), both start at zero.
        # Dynamically match the device and data type of the incoming batch x.
        h0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size, device=x.device, dtype=x.dtype)
        c0 = torch.zeros(self.num_stacked_layers, batch_size, self.hidden_size, device=x.device, dtype=x.dtype)

        out, _ = self.lstm(x, (h0, c0))   # out: hidden state at every timestep
        out = self.fc(out[:, -1, :])       # use only the LAST timestep's hidden state
        return out

model = LSTM(1, 4, 1)   # input_size=1 feature, hidden_size=4, 1 stacked LSTM layer
model.to(device)
model
```

```python
def train_one_epoch():
    model.train(True)          # enable training mode (affects dropout/batchnorm, if any)
    print(f'Epoch: {epoch + 1}')
    running_loss = 0.0

    for batch_index, batch in enumerate(train_loader):
        x_batch, y_batch = batch[0].to(device), batch[1].to(device)

        output = model(x_batch)                    # forward pass
        loss = loss_function(output, y_batch)       # compare prediction vs actual
        running_loss += loss.item()  # .item() detaches the scalar so we don't keep the graph

        optimizer.zero_grad()   # clear gradients from the previous step
        loss.backward()         # backpropagate: compute gradients
        optimizer.step()        # update model weights

        if batch_index % 100 == 99:  # print every 100 batches
            avg_loss_across_batches = running_loss / 100
            print('Batch {0}, Loss: {1:.3f}'.format(batch_index + 1, avg_loss_across_batches))
            running_loss = 0.0

    print()
```

```python
def validate_one_epoch():
    model.train(False)   # switch to evaluation mode
    running_loss = 0.0

    for batch_index, batch in enumerate(test_loader):
        x_batch, y_batch = batch[0].to(device), batch[1].to(device)

        with torch.no_grad():   # no gradients needed for validation -> saves memory/compute
            output = model(x_batch)
            loss = loss_function(output, y_batch)
            running_loss += loss.item()

    avg_loss_across_batches = running_loss / len(test_loader)

    print('Val Loss: {0:.3f}'.format(avg_loss_across_batches))
    print('***************************************************')
    print()
```

```python
learning_rate = 0.001
num_epochs = 10
loss_function = nn.MSELoss()                                   # Mean Squared Error: standard loss for regression
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)  # Adam optimizer updates the weights

for epoch in range(num_epochs):
    train_one_epoch()      # one pass over the training data (updates weights)
    validate_one_epoch()   # one pass over the test data (no weight updates, just monitoring)
```

```python
# Predict on the TRAINING set to visually sanity-check the fit
with torch.no_grad():
    predicted = model(X_train.to(device)).to('cpu').numpy()

plt.plot(y_train, label='Actual Close')
plt.plot(predicted, label='Predicted Close')
plt.xlabel('Day')
plt.ylabel('Close')
plt.legend()
plt.show()
```

```python
train_predictions = predicted.flatten()

# The scaler was fit on 8 columns (Close + 7 lags), so to invert the scaling
# on predictions (which only have 1 column) we rebuild a dummy 8-column
# array, put predictions in column 0, inverse-transform, then take column 0.
dummies = np.zeros((X_train.shape[0], lookback + 1))
dummies[:, 0] = train_predictions
dummies = scaler.inverse_transform(dummies)

train_predictions = dc(dummies[:, 0])   # back to real dollar values
train_predictions
```

```python
# Same trick as above, but for the TEST set predictions
test_predictions = model(X_test.to(device)).to('cpu').detach().numpy().flatten()

dummies = np.zeros((X_test.shape[0], lookback + 1))
dummies[:, 0] = test_predictions
dummies = scaler.inverse_transform(dummies)

test_predictions = dc(dummies[:, 0])
test_predictions
```

```python
# Unscale y_test (the true test-set closing prices) the same way, so the
# comparison plot is in real dollars, not the -1..1 scaled space.
true_dummies = np.zeros((y_test.shape[0], lookback + 1))
true_dummies[:, 0] = y_test.numpy()
true_dummies = scaler.inverse_transform(true_dummies)
unscaled_y_test = dc(true_dummies[:, 0])

# Final plot: predicted vs. actual closing price on the held-out test set
plt.figure(figsize=(10, 5))
plt.plot(unscaled_y_test, label='Actual Close Price')
plt.plot(test_predictions, label='Predicted Close Price')
plt.xlabel('Days (Test Set)')
plt.ylabel('Stock Price')
plt.title('LSTM Future Stock Price Evaluation')
plt.legend()
plt.show()
```
